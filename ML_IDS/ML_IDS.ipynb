```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from sklearn.model_selection import cross_val_score

train = pd.read_csv("C:/Users/sebaa/OneDrive/Studies/Thesis_RISE_2021/Dataset/Training/60min_15min_0.5sec/labelled.csv")
test =  pd.read_csv("C:/Users/sebaa/OneDrive/Studies/Thesis_RISE_2021/Dataset/Predict/30min_10min_0.5sec/testdata.csv")

# Distribution of normal and anomaly traffic in train data
train['traffic'].value_counts()

features = train.columns.tolist()
features.remove('timestamp'), features.remove('traffic')
print(features)

from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(train[features],train['traffic'],train_size=0.70, random_state=1)
print(X_train.shape)
print(X_test.shape)

from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

#Create a svm Classifier
clf1 = svm.SVC(kernel='linear') # Linear Kernel
clf1.fit(train[features], train['traffic'])

#RandomForest classifier
clf2= RandomForestClassifier(n_estimators=100)
clf2.fit(train[features], train['traffic'])

#KNN classifer
clf3 = KNeighborsClassifier(n_jobs=-1)
clf3.fit(train[features], train['traffic'])

from sklearn import metrics

models = []
models.append(('SVM', clf1))
models.append(('RandomForest', clf2))
models.append(('KNN', clf3))

for i, v in models:
    scores = cross_val_score(v, X_train, Y_train, cv=10)
    accuracy = metrics.accuracy_score(Y_train, v.predict(X_train))
    confusion_matrix = metrics.confusion_matrix(Y_train, v.predict(X_train))
    classification = metrics.classification_report(Y_train, v.predict(X_train))
    print()
    print('============================== {} Model Evaluation =============================='.format(i))
    print()
    print ("Cross Validation Mean Score:" "\n", scores.mean())
    print()
    print ("Model Accuracy:" "\n", accuracy)
    print()
    print("Confusion matrix:" "\n", confusion_matrix)
    print()
    print("Classification report:" "\n", classification) 
    print()
    
for i, v in models:
    accuracy = metrics.accuracy_score(Y_test, v.predict(X_test))
    confusion_matrix = metrics.confusion_matrix(Y_test, v.predict(X_test))
    classification = metrics.classification_report(Y_test, v.predict(X_test))
    print()
    print('============================== {} Model Test Results =============================='.format(i))
    print()
    print ("Model Accuracy:" "\n", accuracy)
    print()
    print("Confusion matrix:" "\n", confusion_matrix)
    print()
    print("Classification report:" "\n", classification) 
    print()   
    
  
import plotly.express as px
#Predict for test dataset
#SVM
pred = clf1.predict(test[features])
test['traffic'] = pred
test.to_csv('predicted_svm.csv')

df = pd.read_csv('predicted_svm.csv')
print(df['traffic'].value_counts())
fig = px.line(df, x = 'timestamp', y = 'traffic', title = 'Predictions_SVM')
fig.show()

#RF
pred = clf2.predict(test[features])
test['traffic'] = pred
test.to_csv('predicted_rf.csv')

df = pd.read_csv('predicted_rf.csv')
print(df['traffic'].value_counts())
fig = px.line(df, x = 'timestamp', y = 'traffic', title = 'Predictions_RF')
fig.show()


#KNN
pred = clf3.predict(test[features])
test['traffic'] = pred
test.to_csv('predicted_KNN.csv')

df = pd.read_csv('predicted_KNN.csv')
print(df['traffic'].value_counts())
fig = px.line(df, x = 'timestamp', y = 'traffic', title = 'Predictions_KNN')
fig.show()


```
